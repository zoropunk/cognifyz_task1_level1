```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\ragha\\Downloads\\Dataset .csv")
```

```python
df
```

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
df = df.dropna()
```

```python
display(df.columns)
# Replace 'actual_column_name' with a real column name from your dataset 
df.loc[:, 'Restaurant ID'] = df['Restaurant ID'].fillna(df['Restaurant ID'].median()) 
display(df.isnull().sum())
```

```python
df.dtypes
```

```python
# Ensure you're working with the original DataFrame directly
df.loc[:, 'Restaurant ID'] = pd.to_numeric(df['Restaurant ID'], errors='coerce')
```

```python
df.loc[:, 'Restaurant Name'] = df['Restaurant Name'].astype('category')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], kde=True)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
class_counts = df['Aggregate rating'].value_counts()
print("Class counts for Aggregate rating:")
print(class_counts)

# Optional visualization
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Aggregate rating')
plt.title("Class Imbalance in Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```
